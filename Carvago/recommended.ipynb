```python
df.drop_duplicates(inplace=True)
```

```python
# Afficher les colonnes de chaque DataFrame
print("Colonnes dans df1:", df1.columns.tolist())
print("Colonnes dans df:", df.columns.tolist())

# Comparer les types de données
print("\nTypes de données dans df1:")
print(df1.dtypes)
print("\nTypes de données dans df:")
print(df.dtypes)

# Trouver les différences de colonnes
colonnes_diff = set(df1.columns).symmetric_difference(set(df.columns))
print("\nColonnes différentes:", colonnes_diff)

# Comparer les types pour les colonnes communes
colonnes_communes = set(df1.columns).intersection(set(df.columns))
types_diff = []

for col in colonnes_communes:
    if df1[col].dtype != df[col].dtype:
        types_diff.append({
            'colonne': col,
            'type_df1': df1[col].dtype,
            'type_df': df[col].dtype
        })

print("\nDifférences de types pour les colonnes communes:")
for diff in types_diff:
    print(f"Colonne: {diff['colonne']}")
    print(f"Type dans df1: {diff['type_df1']}")
    print(f"Type dans df: {diff['type_df']}")
```

```python
# Obtenir les types de données de df1 comme référence
df_types = df1.dtypes

# Convertir les colonnes de df pour correspondre à df1, sauf Link
for column in df.columns:
    if column in df1.columns and column != 'Link':
        df[column] = df[column].astype(df1[column].dtype)

# Vérifier la conversion
print("Types après conversion:")
print("\nTypes dans df:")
print(df.dtypes)
print("\nTypes dans df1:")
print(df1.dtypes)

# Vérifier les colonnes
print("\nColonnes dans df1:", df1.columns.tolist())
print("Colonnes dans df:", df.columns.tolist())
```

```python
# Obtenir les types de données de df1 comme référence
df_types = df1.dtypes

# 1. Vérifier les colonnes avec des NA
print("Colonnes avec des NA:")
print(df.isna().sum())

# 2. Nettoyer les NA selon le type de données
# Remplacer NA par 0 pour les colonnes numériques
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].fillna(0)

# 3. Convertir les types (sauf Link)
for column in df.columns:
    if column in df1.columns and column != 'Link':
        df[column] = df[column].fillna(0)  # Assurer qu'il n'y a pas de NA
        df[column] = df[column].astype(df1[column].dtype)

# Vérifier la conversion
print("\nTypes après conversion:")
print(df.dtypes)
print("\nTypes dans df1:")
print(df1.dtypes)

# Vérifier les colonnes
print("\nColonnes dans df1:", df1.columns.tolist())
print("Colonnes dans df:", df.columns.tolist())
```

```python
# 1. Nettoyer les NA et convertir les types
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].fillna(0)

# 2. Convertir tous les types sauf Link pour correspondre à df1
for column in df.columns:
    if column in df1.columns and column != 'Link':
        df[column] = df[column].fillna(0)
        df[column] = df[column].astype(df1[column].dtype)

# 3. Sauvegarder dans recommandation.csv
df.to_csv('recommandation.csv', index=False)

# 4. Vérifier la sauvegarde
print("Vérification des types après sauvegarde:")
df_verification = pd.read_csv('recommandation.csv')
print(df_verification.dtypes)
```

```python
# Lire le fichier
df = pd.read_csv('recommandation.csv')

# Supprimer les doublons
df.drop_duplicates(inplace=True)

# Supprimer les lignes contenant des NaN
df.dropna(inplace=True)

# Sauvegarder le fichier nettoyé
df.to_csv('recommandation.csv', index=False)

# Vérifier le nettoyage
print("Nombre de lignes avant nettoyage:", len(df))
print("Valeurs NaN restantes:", df.isna().sum().sum())
print("Doublons restants:", df.duplicated().sum())
```

```python
# 1. Lire les fichiers
import pandas as pd

df_recom = pd.read_csv('recommandation.csv')
df_cars = pd.read_csv('Cars.csv')

# 2. Obtenir les colonnes communes (sauf Link)
common_columns = [col for col in df_recom.columns if col in df_cars.columns and col != 'Link']

# 3. Trouver les lignes communes
merged_df = pd.merge(
    df_recom, 
    df_cars[common_columns],
    on=common_columns,
    how='inner'
)

# 4. Sauvegarder le résultat
merged_df.to_csv('iliass.csv', index=False)

# 5. Afficher les statistiques
print(f"Nombre de lignes dans recommandation.csv: {len(df_recom)}")
print(f"Nombre de lignes dans Cars.csv: {len(df_cars)}")
print(f"Nombre de lignes communes (dans iliass.csv): {len(merged_df)}")
```

```python
# Identifier les colonnes numériques
numeric_columns = df_cars.select_dtypes(include=['int64', 'float64']).columns

# Calculer et afficher les min/max pour chaque colonne numérique
stats_df = pd.DataFrame({
    'Colonne': numeric_columns,
    'Minimum': [df_cars[col].min() for col in numeric_columns],
    'Maximum': [df_cars[col].max() for col in numeric_columns]
})

# Afficher les résultats
print("Statistiques des colonnes numériques:")
print(stats_df.to_string(index=False))

# Option alternative avec description plus détaillée
print("\nDescription détaillée:")
print(df_cars[numeric_columns].describe())
```
